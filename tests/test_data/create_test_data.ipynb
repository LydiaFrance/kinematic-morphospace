# Create Testing Data

> 2024-06-03 

This provides just five sequences, one for each bird, perch distance, with and without obstacle, with and without IMU, some experienced and some naive, and one with a left turn. 

```python
# Create a very short version of the data for testing
full_file = pd.read_csv("../../data/2024-03-24-FullUnilateralMarkers.csv")

#  01_09_017_2  Drogon, 2020, No IMU, No Obstacle, Experienced
#  02_07_074    Rhaegal, 2017, 7m, Naive
#  03_05_002    Ruby, 2017, 5m, Experienced
#  04_12_268    Toothless, 2017, 12m, Naive
#  05_09_057_1  Charmander, 2020, IMU, Obstacle, Naive, Left Turn


seqList = ['01_09_017_2', '02_07_074', '03_05_002', '04_12_268', '05_09_057_1']

# Create a very short version of the data for testing by getting one sequence per bird
testFile = full_file[(full_file['seqID'].isin(seqList))]

obstacle_df = pd.read_csv('../../data/2024-06-01-ObstacleTurnsSeqList.csv')
testFile = testFile.merge(obstacle_df, how='left', on='seqID')

print(testFile)

# Save to a new csv file
# Get today's date
today = date.today()
testFile.to_csv(str(today)+'-TestUnilateralMarkers.csv', index=False)

```
